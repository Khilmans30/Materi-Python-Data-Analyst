```python
# import libraries

#data wrangling
import numpy as np
import pandas as pd

#data visualisation
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataof air passangers dataset
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url)
```

```python
# Display the first few rows of the dataset
data.head()
```

```python
# check Info
data.info()
```

```python
# transform datatype with pd to datetime
```

```python
data['Month'] = pd.to_datetime(data['Month'])
```

```python
# create a basic line chart for air passenger data
plt.figure(figsize=(20,5))
plt.plot(data['Month'],data['Passengers'],label = 'Passenger Count', color='c',marker='v')
plt.title('International airline passenger overtime')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45,ha='center')
plt.legend()
plt.grid()
plt.show()
```

```python
# create a basic line chart for air passenger data
plt.figure(figsize=(20,5))
sns.lineplot(data=data,x='Month',y='Passengers',label = 'Passenger Count', color='green',marker='v')
plt.title('International airline passenger overtime')
plt.xlabel('Month')
plt.ylabel('Passenger Count')
plt.xticks(rotation=45,ha='center')
plt.legend()
plt.grid()
plt.show()
```

```python
# Load the tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# create a basic vertical bar chart for day-wise tip totals
plt.figure(figsize=(7,4))
plt.bar(tips['day'],tips['tip'])
plt.title('Total Tips by day')
plt.xlabel('Day')
plt.ylabel('Total tips')
plt.grid(axis='y', linestyle='--',alpha =0.6)
plt.show()
```

```python
# group data
time_tip_totals = tips.groupby('time')['tip'].sum().reset_index()
```

Visualization with pandas()


```python
# Create a stacked bar chart for time-wise tip totals with pandas
# Create a stacked bar chart for time-wise tip totals with pandas
plt.figure(figsize=(6,4))
ax=time_tip_totals.plot(kind='bar',x='time',y='tip', stacked=True,colormap='viridis')
plt.title('Total Tips by Time Of Day')
plt.xlabel('Time of day')
plt.ylabel('Total tips')
plt.grid(axis='y', linestyle='--',alpha =0.6)
plt.show()
```

Create histogram

to cretae a histogram mathplotlib, use the hist() funtion

```python
# costumizing mathplotlib with histogram
plt.figure(figsize=(7,4))
plt.hist(tips['total_bill'],color='green',alpha=0.8,edgecolor='black')
plt.title('sitribution of total bill amount')
plt.xlabel('total bill amount')
plt.ylabel('frequency')
plt.grid(axis='y', linestyle='--',alpha =0.6)
plt.show
```

we can also use sea born


```python
# create a histogram for the distribution of total bill amount
plt.figure(figsize=(7,4))
ax= sns.histplot(data=tips,x='total_bill',kde='True',color='green',)
ax.lines[0].set_color('red')                                                                                    
plt.title('ditribution of total bill amount')
plt.xlabel('total bill amount')
plt.ylabel('frequency')
plt.grid(axis='y', linestyle='--',alpha =0.6)

plt.show
```

# scatterplot

```python
# costumizing the mathplotlib scatterplot appereance
plt.figure(figsize=(8,4))
sns.scatterplot(data=tips,x='total_bill',y='tip',color='green',marker='o',)
plt.title('scatterplot of total bill vs tip')
plt.xlabel('total bill amount')
plt.ylabel('tip ammount')
plt.grid()

plt.show
```

```python
# using redplot
plt.figure(figsize=(8,4))
sns.regplot(data=tips,x='total_bill',y='tip',color='green',line_kws=dict(color='r'))                                                                                  
plt.title('reg plotof total bill vs tip')
plt.xlabel('total bill amount')
plt.ylabel('tip amount')
plt.grid(True, linestyle='--',alpha =0.6)

plt.show
```
