> ## How To Handling Missing Values
>
>There are 2 major ways we can handle missing values once we are sure.
>
>>1.Deletion (just for Missing Completly at Random/MCAR):
>
>Just simply removing those particular rows whose features are null or value is missing .
>
>Better for larger datasets as it does not affect the dataset overall.
>
>Not recommended for smaller datasets.
>
>
>>2.Imputation(just for missing at rendom and missing not at random):
>
> Replacing the missing values with some Statistical values(mathematically and logically relevant) that won’t disturb and change the pattern of the dataset.
>
>Generally we consider 3 statistical values in imputation method:
>
>- Mean
>
>- Median
>
>- Mode

### **-Pairwaise**
> Row yang terdapat missing values  di kolom tertentu, akan dihapus semua rownya yang mengandung missing values.

```python
# copy dataset
df3=df.copy
```

> Tahap ini masih sama denga proses handling missing values sebelumnya yaitu adalah mengcopy dataset.Sebenarnya, pada realcase tidak perlu dilakukan tahap ini tidak apa-apa.Akan tetapi, karena akan mendemonstrasikan 4 cara handling missing values maka kita akan menggunakan data copyan untuk tiap metode/cara handling missing values dan tetap dataset asli utuh.

```python
# dropping entire column(car)
df3_dropped = df3.dropna(subset=['Car'])
```

> Pada langkah ini saya menghapus missing values menggunakan pairwise method.Sebenarnya, matode pairwise itu hampir sama dengan listwise, akan tetapi metode listwise menghapus row yang missing values disemua kolom sedangkan metode pairwise menghapus row yang ada missing valuesnya pada kolom tertentu atau sesuai yang kita pilih.Pada step ini, saya mencontohkan kolom 'Car'.

### **-Imput Missing Values**
> Metode ini digunakan jika missing valuesnya itu bertipe missing at random and missing not at random.Dalam pengisiannya kita akan menggunakan mean/median pada data int/float dan mode pada data string.

```python
# copy data
df4 = df.copy()
```

> Tahap ini masih sama denga proses handling missing values sebelumnya yaitu adalah mengcopy dataset.Sebenarnya, pada realcase tidak perlu dilakukan tahap ini tidak apa-apa.Akan tetapi, karena akan mendemonstrasikan 4 cara handling missing values maka kita akan menggunakan data copyan untuk tiap metode/cara handling missing values dan tetap dataset asli utuh.

```python
# show descriptive stats (building Area)
df4['BuildingArea'].describe()
```

> karena metode imputation ini kita memerlukan nilai mean,median,dan modus, maka kita perlu mencari nilainya dulu.Maka, didapat nilai mean,median pada kolom 'BuildingArea'.

```python
# fill / impute data in column 'BuildingArea' with mean
df4['BuildingArea']=df4['BuildingArea'].fillna(round(df4['BuildingArea'].mean(),2))
```

```python
# check data afte imputing
df4.isna().sum()
```

```python
# fill every missing value with mean
df4=df4.fillna(df4.mean())
```

> Berikut adalah terlaksananya missing values yang dihandling menggunakan imputation mean pada setiap kolom yang terdapat pada missing valuesnya.

```python
# show data after fill
df4.isna().sum()
```

```python
# feature 'councilarea' still null , since its categorical data -> string
# we can fill this one with mode
df4['CouncilArea'].describe()
```

> Tahap ini kita harus mengetahui berapa nilai mode dari kolom 'CouncilArea' karena type datanya adalah string dan terdapat missing values pada kolom tersebut.Nilai modenya adalah 'Moreland'.

```python
# fill mith moreland
df4['CouncilArea']=df4['CouncilArea'].fillna("moreland")
```

> Pengisian nilai mode 'Moreland' pada kolom 'CouncilArea'

```python
df4.isna().sum()
```

> Setelah dilakukan handling null values menggunakan imputation maka dapat dilihat tidak terdapat missing values pada setiap kolomnya.
